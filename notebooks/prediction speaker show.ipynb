```python
X = None
for path, fields in features.iteritems():
    field, description = itertools.izip(*fields)
    d = pandas.read_table(path, delim_whitespace=True, header=None, index_col=0) 
    d.columns = field
    if X is None:
        X = d
    else:
        X = X.join(d)
X.index.names = ['SHOW#FirstName_LASTNAME']
```
