### Family of features

Because some features are strongly correlated, we try to automatically derive family of features.
The idea is to perform clustering based on the correlation similarity matrix.

```python
correlation = abs(np.corrcoef(X.T))
imshow(correlation, interpolation='nearest', cmap=cm.Blues); colorbar();
```

Here are the families obtained using DBSCAN clustering algorithm (chosen because it does not require to provide the number of clusters a priori).

```python
from sklearn.cluster import DBSCAN
from pprint import pprint
clustering = DBSCAN(metric='precomputed', min_samples=1)
clusters = clustering.fit_predict(1. - correlation)

families = {}
_families = {}
names = X.columns
for i, cluster in enumerate(clusters):
    if cluster != -1:
        _families[cluster] = _families.get(cluster, []) + [i]
        families[cluster] = families.get(cluster, []) + [names[i]]
for family in families.values():
    print family
```

We keep the most central feature in each family (i.e. the one with the highest average correlation with the other members of the family).

```python
representants = []
for _family in _families.values():
    if len(_family) > 1:
        representant = X.columns[_family[np.argmax(np.mean(np.take(np.take(correlation, _family, axis=0), _family, axis=1), axis=1))]]
    else:
        representant = X.columns[_family[0]]
    representants.append(representant)
representants
```

Let'us re-run the experiments but with family representants only.

```python
sX = X[representants]
```

Classification "Never recognized vs. recognized at least once"

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(max_depth=3).fit(sX, y > 0)
sorted(list(itertools.izip(sX.columns, tree.feature_importances_)), key=lambda x: x[1], reverse=True)
```
